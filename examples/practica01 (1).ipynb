```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Nuevos datos con 16,696 votos totales y porcentajes actualizados
votos_totales = 16696
porcentaje_aldoier = 47.6 / 100  # Convertido a decimal
porcentaje_syda = 52.4 / 100  # Actualizado a 52.4%

# Votos actuales para cada opción
votos_aldoier = votos_totales * porcentaje_aldoier
votos_syda = votos_totales * porcentaje_syda

# Proyección de votos en base al mismo porcentaje
horas_restantes = 2 + (52 / 60)  # 2 horas y 52 minutos convertidos a horas
ritmo_estimado_por_hora = votos_totales / 20  # Aproximación para el ritmo de votos por hora
votos_totales_proyectados = votos_totales + (ritmo_estimado_por_hora * horas_restantes)

# Votos proyectados para cada opción
votos_aldoier_proyectados = votos_totales_proyectados * porcentaje_aldoier
votos_syda_proyectados = votos_totales_proyectados * porcentaje_syda

# Crear DataFrame con los datos
datos = {
    "Fecha": ["Actual", "Proyectado", "Actual", "Proyectado"],
    "Votos": [votos_aldoier, votos_aldoier_proyectados, votos_syda, votos_syda_proyectados],
    "Opciones": ["Aldoier", "Aldoier", "Syda", "Syda"]
}

df = pd.DataFrame(datos)

# Calcular la diferencia y el margen de error para ambos ships (Aldoier y Syda)
diferencia_aldoier = votos_aldoier_proyectados - votos_aldoier
diferencia_syda = votos_syda_proyectados - votos_syda

# Calcular el margen de error en porcentaje
margen_error_aldoier = (diferencia_aldoier / votos_aldoier) * 100
margen_error_syda = (diferencia_syda / votos_syda) * 100

# Mostrar las estadísticas para ambos ships
print(f"**Aldoier**")
print(f"Votos Actuales: {votos_aldoier}")
print(f"Votos Proyectados: {votos_aldoier_proyectados}")
print(f"Diferencia Aldoier: {diferencia_aldoier} votos")
print(f"Margen de Error Aldoier: {margen_error_aldoier:.2f}%\n")

print(f"**Syda**")
print(f"Votos Actuales: {votos_syda}")
print(f"Votos Proyectados: {votos_syda_proyectados}")
print(f"Diferencia Syda: {diferencia_syda} votos")
print(f"Margen de Error Syda: {margen_error_syda:.2f}%\n")

# Configurar estilo de Seaborn
sns.set_theme(style="whitegrid")

# Crear gráfico de líneas
plt.figure(figsize=(10, 6))

# Graficar las votaciones reales y proyectadas para ambas opciones (Aldoier y Syda)
sns.lineplot(data=df, x="Fecha", y="Votos", hue="Opciones", marker="o", palette=["skyblue", "salmon"])

# Añadir título y etiquetas
plt.title("Comparación de votos actuales y proyectados", fontsize=14)
plt.xlabel('Estado de Voto', fontsize=12)
plt.ylabel('Votos', fontsize=12)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# Importar librerías
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos iniciales actualizados
tiempo_restante_horas = 14 + 48 / 60  # Tiempo restante en horas (14 horas y 48 minutos)
votos_totales = 15238  # Total de votos actuales
syda_porcentaje = 55.7  # Porcentaje de Syda
luckity_porcentaje = 44.3  # Porcentaje de Luckity

# Convertir porcentajes en votos actuales
syda_votos = votos_totales * (syda_porcentaje / 100)
luckity_votos = votos_totales * (luckity_porcentaje / 100)

# Datos simulados de evolución histórica (pueden reemplazarse con datos reales)
horas = np.array([0, 5, 10])  # Horas transcurridas (ejemplo)
syda_votos_hist = np.array([5000, 7900, 8486])  # Votos históricos de Syda
luckity_votos_hist = np.array([3913, 5800, 6700])  # Votos históricos de Luckity

# Preparar datos para regresión
X = horas.reshape(-1, 1)  # Tiempo como variable independiente
y_syda = syda_votos_hist  # Votos de Syda como variable dependiente
y_luckity = luckity_votos_hist  # Votos de Luckity como variable dependiente

# Modelos de regresión lineal
model_syda = LinearRegression().fit(X, y_syda)
model_luckity = LinearRegression().fit(X, y_luckity)

# Predicción para el tiempo restante
horas_pred = np.arange(0, tiempo_restante_horas + 1, 1)  # Intervalos de una hora
pred_syda = model_syda.predict(horas_pred.reshape(-1, 1))
pred_luckity = model_luckity.predict(horas_pred.reshape(-1, 1))

# Gráfico
plt.figure(figsize=(12, 6))
plt.plot(horas, syda_votos_hist, label="Syda - Votos Reales", marker='o', color='blue')
plt.plot(horas, luckity_votos_hist, label="Luckity - Votos Reales", marker='o', color='orange')
plt.plot(horas_pred, pred_syda, label="Syda - Predicción", linestyle='--', color='blue')
plt.plot(horas_pred, pred_luckity, label="Luckity - Predicción", linestyle='--', color='orange')
plt.axhline(y=syda_votos, color='blue', linestyle=':', label="Syda - Actual")
plt.axhline(y=luckity_votos, color='orange', linestyle=':', label="Luckity - Actual")
plt.xlabel("Horas transcurridas")
plt.ylabel("Votos")
plt.title("Evolución y Predicción de Votos por Ship")
plt.legend()
plt.grid(True)
plt.show()

# Diferencias actuales
diferencia_porcentual = abs(syda_porcentaje - luckity_porcentaje)
diferencia_votos = abs(syda_votos - luckity_votos)

print(f"Diferencia actual en porcentaje: {diferencia_porcentual:.2f}%")
print(f"Diferencia actual en votos: {diferencia_votos:.0f} votos")

# Evaluar posibilidad de cambio
if any(pred_luckity > pred_syda):
    print("Existe una posibilidad de que Luckity supere a Syda.")
else:
    print("No parece posible que Luckity supere a Syda con base en la tendencia actual.")
```
