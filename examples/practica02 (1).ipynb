```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos iniciales
tiempo_restante_horas = 6 + 1 / 60  # 6 horas y 1 minuto restantes
votos_totales = 37045  # Total de votos actuales
syda_porcentaje = 52.3  # Porcentaje de Syda
shadnap_porcentaje = 47.7  # Porcentaje de Shadnap

# Convertir porcentajes en votos actuales
syda_votos = votos_totales * (syda_porcentaje / 100)
shadnap_votos = votos_totales * (shadnap_porcentaje / 100)

# Datos simulados de evolución histórica (estos pueden reemplazarse por datos reales)
horas = np.array([0, 1, 2, 3, 4, 5])  # Horas transcurridas (ejemplo)
syda_votos_hist = np.array([18000, 22000, 24000, 26000, 28000, 30000])  # Votos históricos de Syda
shadnap_votos_hist = votos_totales - syda_votos_hist  # Votos históricos de Shadnap (complementario)

# Preparar datos para regresión
X = horas.reshape(-1, 1)  # Tiempo como variable independiente
y_syda = syda_votos_hist  # Votos de Syda como variable dependiente
y_shadnap = shadnap_votos_hist  # Votos de Shadnap como variable dependiente

# Modelos de regresión lineal
model_syda = LinearRegression().fit(X, y_syda)
model_shadnap = LinearRegression().fit(X, y_shadnap)

# Predicción para el tiempo restante
horas_pred = np.arange(0, tiempo_restante_horas + 1, 1)  # Intervalos de una hora
pred_syda = model_syda.predict(horas_pred.reshape(-1, 1))
pred_shadnap = model_shadnap.predict(horas_pred.reshape(-1, 1))

# Gráfico
plt.figure(figsize=(12, 6))
plt.plot(horas, syda_votos_hist, label="Syda - Votos Reales", marker='o', color='blue')
plt.plot(horas, shadnap_votos_hist, label="Shadnap - Votos Reales", marker='o', color='orange')
plt.plot(horas_pred, pred_syda, label="Syda - Predicción", linestyle='--', color='blue')
plt.plot(horas_pred, pred_shadnap, label="Shadnap - Predicción", linestyle='--', color='orange')
plt.axhline(y=syda_votos, color='blue', linestyle=':', label="Syda - Actual")
plt.axhline(y=shadnap_votos, color='orange', linestyle=':', label="Shadnap - Actual")
plt.xlabel("Horas transcurridas")
plt.ylabel("Votos")
plt.title("Evolución y Predicción de Votos por Ship")
plt.legend()
plt.grid(True)
plt.show()

# Diferencias actuales
diferencia_porcentual = abs(syda_porcentaje - shadnap_porcentaje)
diferencia_votos = abs(syda_votos - shadnap_votos)
print(f"===== Resumen Actual =====")
print(f"Syda: {syda_porcentaje:.1f}% ({int(syda_votos)} votos)")
print(f"Shadnap: {shadnap_porcentaje:.1f}% ({int(shadnap_votos)} votos)")
print(f"Diferencia actual en porcentaje: {diferencia_porcentual:.2f}%")
print(f"Diferencia actual en votos: {diferencia_votos:.0f} votos")

# Evaluar posibilidad de cambio
if any(pred_shadnap > pred_syda):
    print("Existe una posibilidad de que Shadnap supere a Syda.")
else:
    print("No parece posible que Shadnap supere a Syda con base en la tendencia actual.")
```
