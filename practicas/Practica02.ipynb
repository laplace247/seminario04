```python
#Ejercicio 02
#Crear un DataFrame y agregar una nueva columna calculada
import pandas as pd
import numpy as np

# Configuración para reproducibilidad
np.random.seed(0)

# Generar un DataFrame con valores aleatorios entre 1 y 100
data = {
    'X': np.random.randint(1, 101, size=5),
    'Y': np.random.randint(1, 101, size=5),
    'Z': np.random.randint(1, 101, size=5)
}
df = pd.DataFrame(data)

# Mostrar el DataFrame original
print("DataFrame original:")
print(df)

# Filtrar las filas donde la suma de las columnas X y Y sea mayor a 120
filtrado_df = df[df['X'] + df['Y'] > 120]

# Mostrar el DataFrame filtrado
print("\nDataFrame filtrado:")
print(filtrado_df)
```

```python
#Ejercicio 04
import pandas as pd
import numpy as np

# Establecer una semilla para reproducibilidad
np.random.seed(42)

# Crear dos DataFrames con valores enteros aleatorios entre 10 y 50
df1 = pd.DataFrame(np.random.randint(10, 51, size=(4, 3)), columns=['A', 'B', 'C'])
df2 = pd.DataFrame(np.random.randint(10, 51, size=(4, 3)), columns=['A', 'B', 'C'])

# Mostrar los DataFrames generados
print("DataFrame df1:")
print(df1)
print("\nDataFrame df2:")
print(df2)

# Hacer la resta df1 - df2
resultado = df1 - df2

# DataFrame con resta
print("\nResta:")
print(resultado)

```

```python
#Ejercicio 07
#Reemplazar valores faltantes en un DataFrame
import numpy as np

# Crear DataFrame y agregar algunos NaN
df = pd.DataFrame(np.random.randint(10, 91, size=(5, 3)), columns=['A', 'B', 'C'])
df.iloc[[1, 3, 4], [0, 1, 2]] = np.nan

print("DataFrame con NaNs:")
print(df)

# Reemplazar NaNs por el promedio de cada columna
df.fillna(df.mean(), inplace=True)

print("\nDataFrame con NaNs reemplazados por el promedio de la columna:")
print(df)

```

```python
# Ejercicio 05
# Agrupar y contar los elementos de cada grupo
# Importación de la librería Pandas
import pandas as pd

# Crear el DataFrame
# Creacion del diccionario con los datos
df = pd.DataFrame({
  'Producto': ['Manzana', 'Banana', 'Cereza', 'Durazno', 'Manzana', 'Banana',
                'Cereza', 'Durazno', 'Manzana', 'Banana'],
  'Mes': ['Enero', 'Enero', 'Febrero', 'Febrero', 'Marzo', 'Marzo', 'Abril',
          'Abril', 'Mayo', 'Mayo']
})

# Agrupar por "Productos" y contar cuántas veces aparece cada producto
productos = df.groupby('Producto').size()

# Mostrar el resultado de la operacion agrupado haciendo el conteo de veces
print(productos)
```

```python
# Importar la libreria pandas y renombrarla a pd
import pandas as pd

# Crear el DataFrame con los datos proporcionados en el ejercicio
# Aqui creamos un diccionario con los datos
data = {
    'Producto': ['Lavadora', 'Televisor', 'Refrigeradora', 'Microondas', 'Aspiradora'],
    'Precio': [300, 400, 150, 200, 350],
    'Unidades': [3, 2, 4, 1, 5]
}
# Con esto convertimos el diccionario en tablas donde se dividiran en filas y columnas
df = pd.DataFrame(data)

# Agregar la columna 'Total' calculada por Producto * Unidades
df['Total'] = df['Precio'] * df['Unidades']

# Mostrar el DataFrame resultante de la operacion
print(df)
```

```python
import pandas as pd

# Crear el DataFrame con los datos proporcionados
data = {
    'Producto': ['Manzana', 'Banana', 'Cereza', 'Durazno', 'Manzana', 'Banana',
                 'Cereza', 'Durazno', 'Manzana', 'Banana'],
    'Mes': ['Enero', 'Enero', 'Febrero', 'Febrero', 'Marzo', 'Marzo', 'Abril', 'Abril',
            'Mayo', 'Mayo']
}

df = pd.DataFrame(data)

# Mostrar el DataFrame original
print("DataFrame original:")
print(df)
print("\n")

# Agrupar por 'Producto' y contar las ocurrencias
conteo_productos = df['Producto'].value_counts().reset_index()
conteo_productos.columns = ['Producto', 'Conteo']

# Mostrar el resultado del agrupamiento
print("Conteo de productos:")
print(conteo_productos)
print("\n")

# Información adicional
print("Distribución de productos por mes:")
print(df.groupby('Mes')['Producto'].value_counts().unstack(fill_value=0))
```
