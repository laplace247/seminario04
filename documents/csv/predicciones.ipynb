```python
import numpy as np
import pandas as pd

#guardar semilla
np.random.seed(42)

#generar daos ventas

def generar_datos_venta(nombre_empresa, productos, start, end):
  meses = pd.date_range(start=start,end=end,freq='M')
  datos_ventas = pd.DataFrame({
      'Meses':meses,
      **{producto:np.random.randint(1000,5000,size=len(meses)) for producto in productos}

         })

  #guardar el dataframe como csv
  datos_ventas.to_csv(f'{nombre_empresa}.csv', index=False)

#definir nombre de las empresas
empresas = {
    'Heladeria':['Chocolate','Fresa','Vainilla','Mango'],
    'Restaurante':['Pasta','Pizza','Ensalada','Sushi'],
    'Clothing_store':['Camisas','Pantalones','Vestidos','Blusas'],
    'Librería':['Ficcion','No Ficcion','Historia','Cultura']
}

#generar los datos y guardarlos como csv
for nombre_empresa,productos in empresas.items():
  generar_datos_venta(nombre_empresa,productos,'2020-01-01','2024-09-30')
```

```python
# Analisis de datos
# Importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos de ventas de la tienda de ropa
clothing_store_df = pd.read_csv('Clothing_store.csv')

# Mostrar las primeras filas del dataframe
print(clothing_store_df.head())

promedio_ventas_clothing_store = clothing_store_df[['Camisas', 'Pantalones', 'Vestidos', 'Blusas']].mean()
# Mostrar los resultados
print('Promedio de ventas por tipo de ropa:')
print(promedio_ventas_clothing_store)

# Identificar el producto con mayor venta
producto_recomendado = promedio_ventas_clothing_store.idxmax()
print("\n")
print(f'Se recomienda invertir más en el producto: {producto_recomendado} para octubre del 2024')

# Graficar las ventas promedio por tipo de producto
plt.figure(figsize=(8,6))
promedio_ventas_clothing_store.plot(kind='bar', color=['brown', 'pink', 'blue', 'orange'])
plt.title('Promedio de ventas por producto en tienda de ropa')
plt.xlabel('Tipo de producto')
plt.ylabel('Promedio de ventas')
plt.xticks(rotation=0)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Cargar los datos de la heladería
clothing_store_df = pd.read_csv('Clothing_store.csv')

# Seleccionar solo las columnas de ventas
X = clothing_store_df[['Camisas','Pantalones','Vestidos','Blusas']].values

# Escalar los datos entre 0 y 1
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Crear las etiquetas (Y) con los mismos datos de las ventas, desplazado en el tiempo
Y = np.roll(X_scaled, -1, axis=0)

# Separar los datos de entrenamiento y validación
X_train, Y_train = X_scaled[:-1], Y[:-1]

# Definir el modelo
model = Sequential()
model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(64, activation='relu'))  # Corrección: 'activation' en vez de 'actviation'
model.add(Dense(4, activation='linear'))

# Compilar el modelo
model.compile(optimizer='adam', loss='mean_squared_error')

# Entrenar el modelo
history = model.fit(X_train, Y_train, epochs=200, batch_size=16, verbose=1)

# Predicción para las ventas del 2025
prediccion_2025 = model.predict(X_train[-1].reshape(1, -1))
ventas_2025 = scaler.inverse_transform(prediccion_2025)  # Desescalar los valores

# Mostrar las predicciones de ventas del 2025
productos = ['Camisas','Pantalones','Vestidos','Blusas']
ventas_2025 = ventas_2025.flatten()
for i, producto in enumerate(productos):
    print(f'Predicción de ventas para {producto} en 2025: {ventas_2025[i]:.2f}')

#agregamos grafica
plt.figure(figsize=(8,6))
plt.bar(productos, ventas_2025, color=['brown','pink','beige','orange'] )
plt.title('Promedio de ventas por producto en la tienda de ropa para el 2025')
plt.xlabel('Tipos')
plt.ylabel('Promedio Proyectadas')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Cargar los datos de la heladería
heladeria_df = pd.read_csv('Heladeria.csv')

# Seleccionar solo las columnas de ventas
X = heladeria_df[['Chocolate', 'Fresa', 'Vainilla', 'Mango']].values

# Escalar los datos entre 0 y 1
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Crear las etiquetas (Y) con los mismos datos de las ventas, desplazado en el tiempo
Y = np.roll(X_scaled, -1, axis=0)

# Separar los datos de entrenamiento y validación
X_train, Y_train = X_scaled[:-1], Y[:-1]

# Definir el modelo
model = Sequential()
model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(64, activation='relu'))  # Corrección: 'activation' en vez de 'actviation'
model.add(Dense(4, activation='linear'))

# Compilar el modelo
model.compile(optimizer='adam', loss='mean_squared_error')

# Entrenar el modelo
history = model.fit(X_train, Y_train, epochs=200, batch_size=16, verbose=1)

# Predicción para las ventas del 2025
prediccion_2025 = model.predict(X_train[-1].reshape(1, -1))
ventas_2025 = scaler.inverse_transform(prediccion_2025)  # Desescalar los valores

# Mostrar las predicciones de ventas del 2025
productos = ['Chocolate', 'Fresa', 'Vainilla', 'Mango']
ventas_2025 = ventas_2025.flatten()
for i, producto in enumerate(productos):
    print(f'Predicción de ventas para {producto} en 2025: {ventas_2025[i]:.2f}')

# Predecir las ventas para todo el set de datos de entrenamiento
predicciones = model.predict(X_train)

# Invertir la escala de las predicciones y de los valores reales para tener los datos originales
predicciones_invertidas = scaler.inverse_transform(predicciones)
ventas_reales_invertidas = scaler.inverse_transform(X_train)

# Promedio de ventas proyectadas para cada producto en 2025
promedio_ventas_2025 = predicciones_invertidas.mean(axis=0)

# ----------------------- Gráfico 1: Barras para el promedio de ventas 2025 -----------------------
plt.figure(figsize=(8,6))
plt.bar(productos, promedio_ventas_2025, color=['brown', 'pink', 'blue', 'orange'])
plt.title('Promedio de ventas proyectadas para el 2025')
plt.xlabel('Sabores')
plt.ylabel('Ventas promedio')
plt.xticks(rotation=0)
plt.show()

# ----------------------- Gráfico 2: Comparación de ventas reales vs predicciones -----------------------
# Promedio de ventas históricas
promedio_ventas_historicas = ventas_reales_invertidas.mean(axis=0)

# Crear DataFrame para comparación
df_comparacion = pd.DataFrame({
    'Sabores': productos,
    'Promedio Ventas Históricas': promedio_ventas_historicas,
    'Promedio Predicciones 2025': promedio_ventas_2025
})

# Gráfico de barras para comparar
df_comparacion.plot(x='Sabores', kind='bar', figsize=(10, 6), color=['skyblue', 'orange'])
plt.title('Comparación entre ventas históricas y predicciones para 2025')
plt.ylabel('Ventas promedio')
plt.xticks(rotation=0)
plt.show()

# ----------------------- Gráfico 3: Gráfico de líneas de tendencias -----------------------
# Crear DataFrame con las ventas reales y predicciones
df_tendencias = pd.DataFrame(ventas_reales_invertidas, columns=productos)

# Agregar un último valor que represente las predicciones del 2025
prediccion_2025_extendida = pd.DataFrame([ventas_2025], columns=productos)
df_tendencias = pd.concat([df_tendencias, prediccion_2025_extendida], ignore_index=True)

# Graficar las tendencias de ventas
plt.figure(figsize=(10,6))
for producto in productos:
    plt.plot(df_tendencias.index, df_tendencias[producto], label=f'{producto} Real')
plt.title('Tendencias de ventas históricas y predicciones para 2025')
plt.xlabel('Tiempo')
plt.ylabel('Ventas')
plt.legend()
plt.show()

# ----------------------- Gráfico 4: Heatmap de correlación entre sabores -----------------------
# Calcular la matriz de correlación
corr_matrix = pd.DataFrame(ventas_reales_invertidas, columns=productos).corr()

# Crear el heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlación entre ventas de diferentes sabores')
plt.show()

# ----------------------- Gráfico 5: Dispersión de predicciones vs reales -----------------------
plt.figure(figsize=(8,6))
for i, producto in enumerate(productos):
    plt.scatter(ventas_reales_invertidas[:, i], predicciones_invertidas[:, i], label=producto)
plt.title('Comparación de ventas reales vs predicciones')
plt.xlabel('Ventas Reales')
plt.ylabel('Ventas Predichas')
plt.legend()
plt.show()
```
